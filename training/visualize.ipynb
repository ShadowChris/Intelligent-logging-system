```python
import matplotlib.pyplot as plt
import pandas as pd

# 读取数据
data = pd.read_csv('train_loss.txt', sep='\t')

# 提取步骤和训练损失
steps = data['Step']
loss = data['Training Loss']

# 创建曲线图
plt.figure(figsize=(10,6))
plt.plot(steps, loss, label='Training Loss')

# 添加标题和标签
plt.title('Training Loss over Steps')
plt.xlabel('Steps')
plt.ylabel('Training Loss')

# 显示图例
plt.legend()

# 显示图形
plt.show()
```
